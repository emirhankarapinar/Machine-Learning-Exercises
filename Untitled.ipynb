```python
a=pd.read_csv("musteri.csv")
df3=pd.DataFrame(a)
```

```python
x4=df3.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans()
yheadd=kmeans.fit_predict(x4)
wcss=[]
for i in range(1,14):
    kmeans.fit(x4)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,14),(wcss))
```
